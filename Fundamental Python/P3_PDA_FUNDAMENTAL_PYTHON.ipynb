## **C. Variabel Python**

Variable adalah lokasi memori yang dapat digunakan untuk menyimpan sebuah
nilai yang dapat diubah ubah sesuai kebutuhan. Variabel pada Python bersifat dinamis dan tidak perlu dideklarasikan terlebih dahulu dengan tipe data tertentu. Variabel nya juga dapat diubah saat program dijalankan. (Wahyono, 2018)

Aturan Penamaan variabel pada python :
1. Karakter pertama harus huruf atau garis bawah.
2. Karakter berikutnya yang bisa digunakan adalah huruf, angka, atau garis bawah.
3. Panjang variable tidak terbatas karakter namun gunakan seperlunya saja.
4. Penamaan variable sifatnya case sensitive.
5. Jangan menggunakan keyword python.

# **F. Kondisi (if.., else... elif**
Pada Python pernyataan kondisional menggunakan pola if ... elif ... else . Mirip
dengan bahasa C/C++, bedanya pada Python tidak menggunakan else if, namun
menggunakan elif. Selain itu pada Python untuk memulai scope baru, tidak
menggunakan tanda kurung kurawal “{}”, tetapi menggunakan indentasi

**Contoh:**


```
a = 200
b = 33
if b > a:
  print("b lebih besar dari a")
elif a == b:
  print("a dan b bernilai sama")
else:
  print("a lebih besar daripada b")

```

## **K. Slicing**
Slicing merupakan teknik memilih data dari tipe data sequence , set tidak dapat
melakukan operasi ini



```
asprak = {"fahmi", "rafelino", "lingga", "sarah"} 
print(asprak)
asprak = ["fahmi", "rafelino", "lingga", "sarah", "sandy"] 
print(asprak)
print(asprak[1]) #rafelino
print(asprak[-1]) #sandy
print(asprak[1:3]) #[‘rafelino’, ‘lingga’]
print(asprak[:2]) #[‘fahmi’, ‘lingga’]
print(asprak[3:]) #[‘sarah’, ‘sandy’]
print(asprak[-2:-1]) #[‘sarah’]
print(asprak[0:5:2]) #[‘fahmi’, ‘bana’, ‘sandy’]
```

## **M. Fungsi dengan parameter**
Pada fungsi yang dapat menerima parameter, fungsi yang memanggil harus
memberikan jumlah argumen yang sama dengan parameter yang dideklarasikan.
Parameter adalah deklarasi variabel pada deklarasi fungsi, sedangkan argumen adalah nilai yang diberikan kepada parameter

````
def nama(depan, belakang = "fiqri"):
  print(depan, belakang)

  nama("fahmi")
  nama("fahmi", "noor")
  nama(depan = "fahmi", belakang="noor") 
````

## **N.Fungsi dengan Nilai Kembalian ( Rekursif**) 
Fungsi dengan nilai kembalian tidak jauh berbeda dengan fungsi pada umumnya.
Perbedaannya pada fungsi dengan nilai kembalian terdapat keyword return. (Fiqri, 2020)

````
def tambah(a,b):
  return a+b

print(tambah(20, 1)) 
````
